# LAB: Descriptive and exploratory data analysis

First of all, we begin by importing the packages we will use in this lab and loading the data from our CSV file into the DataFrame.

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
df = pd.read_csv("/Users/Alejandro/Downloads/Actividad 1/Actividad 1/cancer.csv")
```

### Characterization of the distributions

```python
frec_relativa = df['Uniformity of Cell Size'].value_counts(normalize = True)
print(frec_relativa)
```

```python
plt.hist(df['Uniformity of Cell Size'], bins = 10, edgecolor = 'black')
plt.title("Histogram of Uniformity of Cell Size")
plt.show()
```

```python
media = df['Uniformity of Cell Size'].mean()
mediana = df['Uniformity of Cell Size'].median()
moda = df['Uniformity of Cell Size'].mode()[0]
print("Mean:", media)
print("Median:", mediana)
print("Mode:", moda)
```

```python
## Skewness coefficient. If coefficient > 0: Positive skewness; If coefficient < 0: Negative skewness; If coefficient = 0: Perfectly symmetric.
coe_as = df['Uniformity of Cell Size'].skew()
print("Skewness coefficient:", coe_as)
```

When we calculate the relative frequency, we can observe that almost 55% of our data has a value of 'Uniformity of Cell Size' as one.
Analyzing the histogram, we can see that the distribution **is not** symmetric, and therefore, by analyzing the values of mean, median, mode, and skewness coefficient, we conclude that the distribution is **positively skewed**.

To obtain this result, we rely on the fact that positive skewness tends to occur when the mode is less than or equal to the median, and the median is less than the mean. Additionally, observing the skewness coefficient, if it is greater than 0, it indicates positive skewness.

### Measures of dispersion

```python
varianza = df['Uniformity of Cell Size'].var()
desviacion = df['Uniformity of Cell Size'].std()
print("Variance:", varianza)
print("Standard Deviation:", desviacion)
```

A **variance** of **9.39511** indicates that the values in the dataset are, on average, distant from the mean. Meanwhile, a **standard deviation** of **3.06514** indicates that the data deviates approximately 3.06514 units from the mean.

With these results, we can conclude that the values in this dataset are not extremely concentrated around the mean, but they are also not extremely disperse."

### Outliers

```python
plt.boxplot([df['Uniformity of Cell Size'], df['Single Epithelial Cell Size']], labels = ['Uniformity of Cell Size', 'Single Epithelial Cell Size'])
plt.ylabel("Value")
plt.title("Uniformity of Cell Size and Single Epithelial Cell Size")
plt.show()
```

The yellow horizontal line indicates the median of the data. In these two cases, we can see that the median is positioned at the bottom of the box, indicating that it is relatively low compared to the total range of data. The median is the value that divides the dataset into two equal parts, and as we have seen previously for the variable 'Uniformity of Cell Size,' we know that the value 1 had more than 50% of the data.

Analyzing the boxplot for the first variable 'Uniformity of Cell Size,' we can see that it does not have a lower whisker; the first quartile is around 1, the third quartile is around 5, and there are no outliers.

In the case of the second variable 'Single Epithelial Cell Size,' we observe both lower and upper whiskers indicating non-outlier minimum and maximum values. We also observe that the first quartile has a value of 2 and the third quartile has a value of 4. In this case, we have outliers marked with circles. An outlier is a value that may be extremely high or low in relation to the rest of the data. In this case, we would have outliers between 8 and 10.

### Correlation between variables

```python
corr_matrix = df.corr()
# We create the correlation matrix plot using the Seaborn package.
plt.figure(figsize=(12, 8))
sns.heatmap(corr_matrix, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```

Upon observing the correlation matrix, we can see how the parallel coordinates plot indicated that there is correlation among most of the variables in this dataset. The strongest correlation point is between the variables 'Uniformity of Cell Size' and 'Uniformity of Cell Shape' with a value around 0.9, very close to 1.

Furthermore, we also have a positive correlation between the variables 'Clump Thickness' and 'Uniformity of Cell Size' with a value close to 0.6. Between the variables 'Uniformity of Cell Shape' and 'Marginal Adhesion,' there is also a positive correlation with a value close to 0.6. Similarly, between the variables 'Marginal Adhesion' and 'Single Epithelial Cell Size,' there is a positive correlation with a value close to 0.6. Finally, as expected, we can observe that the variable 'Sample code number' is completely independent and has no correlation with any of the other variables in the dataset.
